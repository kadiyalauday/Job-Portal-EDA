#### Importing the Libraries

``````python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
``````

#### Importing the dataset

``````python
#reading the csv file
df = pd.read_csv("Job_portal.csv")
``````

``````python
#Printing the first 5 rows of data frame
df
``````

**Description about the features**

There are 7 columns in the dataset and the columns are:

    1. Unnamed: 0 : An unnamed or index column, possibly used for row identification.
    2. Company_name : Name of the companies as per the website.
    3. Job_role : Job roles offered by companies.
    4. Min_salary : Minimum salary offered by the company, and the salary is in Indian currency.
    5. Max_salary : Maximum salary offered by the company.
    6. Location : Location of the companies.
    7. Work_mode : Work mode offered by the companies(e.g., work from office, work from home, field job).
    8. Work_type : Work type offered by the companies(e.g., Part time, Full time).

### Data Cleaning

``````python
#Dropping the column because it doesnot contribute any data for the analysis.
df.drop("Unnamed: 0",axis=1,inplace=True)
``````

Removing the special characters

``````python
#Removing the commas from both the columns
``````

``````python
df["Min_salary"]= df["Min_salary"].str.replace(",","")
``````

``````python
df["Max_salary"]= df["Max_salary"].str.replace(",","")
``````

Detecting the duplicate values

``````python
#Checking for the duplicate values and there are 11 duplicate rows.
df.duplicated().sum()
``````

``````python
#Dropping the duplicates 
df.drop_duplicates(inplace=True)
``````

``````python
df.duplicated().sum()
``````

Checking for Missing Values

``````python
#There is no missing values
df.isna().sum()
``````

Type casting

``````python
#Converting object data type into int32 as the data is numerical data.
``````

``````python
df["Min_salary"] = df["Min_salary"].astype("int32")
df["Max_salary"] = df["Max_salary"].astype("int32")
``````

``````python
#Selecting only the numerical columns
numerical = df.select_dtypes(exclude= "object")
``````

``````python
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=1, ncols=len(numerical.columns), figsize=(15, 5))

for i, col in enumerate(numerical.columns):
    sns.kdeplot(numerical[col],fill=True,ax=axes[i])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
``````

Insights from density plot:
* There are extreme values at right side and more data points on the left side so the data is following left skewed distribution.

#### Top 10 job roles with highest minimum salary

``````python
#Sorting the dataframe by Min_salary column in descending order.

sorted_df= df.sort_values(by= "Min_salary",ascending=False)
top_10 = sorted_df.head(10)

#Taking the first 10 rows after sorting
``````

``````python
plt.figure(figsize=(6,4))
sns.barplot(data= top_10, y="Job_role", x="Min_salary",color="#3775fa")
``````

Insights from above barplot:

* Process Design Engineer role is offering highest minimum salary is 149000
* Civil/site engineer positions often offer the highest minimum salaries.
* The black lines in the bars are the error bars which represent variability in the data points.
* Civil engineer, civil site engineer has uncertainty in the data points.
